```python
# Dependencies and Setup
import pandas as pd
import numpy as np 
import os
from scipy.stats import sem
from scipy.stats import linregress
import scipy.stats as stats

# For plotting we add addtional dependencies
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib.colors as colors

# Turn off warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
# Set the color palette
color_taupe = (150/255, 140/255, 140/255)       # Taupe
color_sage = (165/255, 171/255, 129/255)        # Sage
color_rust = (221/255, 128/255, 71/255)         # Rust
color_green = (123/255, 167/255, 157/255)       # Green
color_blue = (148/255, 182/255, 210/255)        # ice Blue
color_brown = (119/255, 95/255, 85/255)         # Brown
color_gold = (216/255, 178/255, 92/255)         # Brown
color_tan = (235/255, 221/255, 195/255)         # Brown
```

```python
# Load the CSV file a Pandas DataFrame
df = pd.read_csv("./Resources/all_data_clean.csv")
```

```python
# box plot of Admission Rates
color_dict = {'flierprops':dict(markeredgecolor = color_rust,                   # set the color of the outlier outline
                                markerfacecolor = color_rust,                   # set the color of the outlier fill
                                marker = 'o',                                   # set the outlier marker shape
                                markersize=10)}

medianprops = dict(linestyle='-', linewidth=2, color=color_rust)

df.boxplot("admin_rate", 
            figsize=(10,10), 
            fontsize = 14, 
            grid = False, 
            medianprops=medianprops, 
            color='black', 
            **color_dict)

plt.title('College Admission Rates',fontsize = 24,fontweight="bold", color=color_rust) 

plt.ylabel('Rate of Acceptance',                # Add a y-axis description
            fontsize = 16, 
            color=color_rust, 
            fontweight="bold")

save_fig = os.path.join('.', 'Output', 'Adm_ratebox.jpg')
plt.savefig(save_fig, bbox_inches='tight')
```

```python
# scatter plot of Admission Rate vs Tuition costs
ax = df.plot.scatter(x="avg_net_price", y="admin_rate",figsize=(16,10),color=color_taupe, fontsize=16)

plt.title("Admission Rate vs. Tuition Costs",color=color_rust, fontsize = 24, fontweight="bold")

plt.ylabel('Admission Rate',                # Add a y-axis description
            fontsize = 16, 
            color=color_rust, 
            fontweight="bold")

plt.xlabel('Tuition Costs',                # Add a x-axis description
            fontsize = 16, 
            color=color_rust, 
            fontweight="bold")

xdata = df['avg_net_price']
ydata = df['admin_rate']  
    
    
    # linear Regression equation calculation
(slope, intercept, rvalue, pvalue, stderr) = linregress(xdata, ydata)
regress_values = xdata * slope + intercept

    # create a string annoation that for the equation of the line and r**2
line_eq_str = "y = " + str(round(slope,5)) + "x + " + str(round(intercept,2))       
r_square_str = f'R-Squared = {round(rvalue**2,2)}'                                  
plt.plot(xdata,regress_values,'red')

print(line_eq_str)
print(r_square_str)

save_fig = os.path.join('.', 'Output', 'Adm_ratescatter.jpg')
plt.savefig(save_fig, bbox_inches='tight')
```

```python
# box plot of SAT Scores

color_dict = {'flierprops':dict(markeredgecolor = color_rust,                   # set the color of the outlier outline
                                markerfacecolor = color_rust,                   # set the color of the outlier fill
                                marker = 'o',                                   # set the outlier marker shape
                                markersize=10)}

df.boxplot("SAT_Avg", 
            figsize=(10,10), 
            fontsize = 14, 
            grid = False, 
            medianprops=medianprops, 
            color='black', 
            **color_dict)

plt.title('SAT Scores',fontsize = 24,fontweight="bold",color=color_rust)

plt.ylabel('Scores',                # Add a y-axis description
            fontsize = 16, 
            color=color_rust, 
            fontweight="bold")

save_fig = os.path.join('.', 'Output', 'SATbox.jpg')
plt.savefig(save_fig, bbox_inches='tight')
```
